```python
from collections import Counter
```

```python
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
```

```python
histogram = Counter(min((grade//10)*10 ,90) for grade in grades)
plt.bar([x+5 for x in histogram.keys()],histogram.values(),10,edgecolor=(0,0,0))
#x+5 means center should be at 15 else center will come at 0,10, ... , bar_width = 10
plt.axis([-5,105,0,5])
#(x1,x2,y1,y2)
plt.xticks([10*i for i in range(11)])
plt.xlabel("Decile")
plt.ylabel("# of students")
plt.title("Distribution of exam grades")
plt.show()
```

```python
mentions = [500,505]
years = [2017,2018]

plt.bar(years, mentions, 0.8)
plt.xticks(years)
plt.ylabel("# of time DS words taken")
plt.ticklabel_format(useOffset = False)
plt.axis([2016.5,2018.5,499,506])
plt.title("Look at huge increase")
plt.show()
```

```python
plt.bar(years, mentions, 0.8)
plt.xticks(years)
plt.ylabel("# of time DS words taken")
plt.axis([2016.5,2018.5,0,550])
plt.title("Look at not so huge increase")
plt.show()
```

```python
variance     = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [x+y for x,y in zip(variance,bias_squared)]
xs = [i for i,_ in enumerate(variance)]
```

```python
plt.plot(xs, variance, 'g-', label="variance")
plt.plot(xs, bias_squared, 'r-.', label="bias_squared")
plt.plot(xs, total_error, 'b:', label="total_error")
plt.legend(loc=9)
plt.xlabel("model_complexity")
plt.xticks([])
plt.title("The bias variance tradeoff")
plt.show()
```

```python
friends = [ 70,  65,  72,  63,  71,  64,  60,  64,  67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels =  ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
```

```python
plt.scatter(friends,minutes)
for label, friend_count, minute_spend in zip(labels,friends,minutes):
    plt.annotate(label, xy=(friend_count,minute_spend),xytext=(-5,5),textcoords="offset points")
plt.title("Daily minutes vs #of friends")
plt.xlabel("# of friends")
plt.ylabel("Minutes spent")
plt.show()
```
